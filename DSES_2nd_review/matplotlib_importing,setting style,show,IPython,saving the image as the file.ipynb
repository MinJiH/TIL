# Visualization using Matplotlib

Matplotlib은 넘파이 배열을 기반으로 만들어진 다중 플랫폼 데이터 시각화 라이브러리로서 광범위한 사이파이 스택과 함께 작업하기 위해 설계됨.
다양한 운영체제와 그래픽 백엔드에서 잘 동작한다는 장점이 있음.


# Normal Matplotlib method

**1. importing matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**2.Setting Style**

```python
plt.style.use('classic') # this is for 'classic style'.. for more, https://matplotlib.org/gallery/style_sheets/style_sheets_reference.html
```

**3.show()**
* finding all the object for image,opening the new window for new image

```python
x = np.linspace(0,10,100) #  Return evenly spaced numbers over a specified interval(in this case, from 0 to 10, creating spaced numbers)
display(x)
```

```python
plt.plot(x,np.sin(x))
```

```python
x = np.linspace(0,10,100)
display(x)

# drawing function cos, sin
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

# plotting means creating the point on coordinate plane

plt.show(0)
```

**4.%matplot**
 * in IPython shell, call the "%matplotlib"command in order to use the matplotlib as conversational express
 * in this mode, no need to call 'show()' function
 

```python
%matplotlib
```

```python
%matplotlib inline 
# this is the line for expressing the graph on jupyter notebook
# inserting the image of plot into the notebook
# only one callling is enough for one kernnel
```

```python
x = np. linspace(0,10,100)

# this is for saving the plot as the file
fig = plt.figure()

plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
```

**5. Saving the image as file **

```python
fig.savefig('my_figure.png')
```
